```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## 64613
### Hot Water Heater

```python
house_id = 64613
df = pd.read_csv('../data/'+str(house_id)+'.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)

# Hot water Circuit id: 70132
load = df.loc[df.circuit_id == 70132].copy(deep=True)
print("LOAD: " + load.monitors.iloc[0])

load['reading_datetime'] = pd.to_datetime(load['reading_datetime'])
load = load.set_index('reading_datetime')
load.sort_index(inplace=True)

load['power'] = (load.energy_pos * 12) / 1000
rated_power = load.power.max()
load['time'] = (load.power / rated_power) * 5
print('Rated Power: ' + "{:.2f}".format(rated_power) + 'kW')

load_24h = load[['power', 'time']]
load_24h = load_24h.resample('24h', offset=12).sum()
load_24h.time.plot()
plt.show()
longest_day = load_24h.time.idxmax()

load.loc[str(longest_day.date())].power.plot()
plt.show()
```
