```python
import pandas as pd
import os
import thesisInitial.tariff as tf
from datetime import time, date
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
cwd = os.path.abspath('.')
fp = os.path.join(cwd, "../data/64613.csv")
data = pd.read_csv(fp)
# data['reading_datetime'] = pd.to_datetime(data['reading_datetime'])
data = data.drop(columns=data.columns[0])
data = data.set_index('reading_datetime')
data.index = pd.to_datetime(data.index)
```

```python
load_net = data.loc[data['monitors'] == 'ac_load_net'].copy(deep=True)
```

```python
# Origin Go Tariff
peak        = {(6,10): 40.08, (15,24): 40.08, (0,1): 40.08}
shoulder    = {(10,15): 20.06}
off_peak    = {(1,6): 21.62}

tou_rate = {**peak, **shoulder, **off_peak}

OriginGO = tf.TOUTariff(73.71, 8, tou_rate)
```

```python
cost_kWh = pd.cut(load_net.index.hour,
                  bins=OriginGO.time_bin,
                  include_lowest=True,
                  labels=OriginGO.tariff_bin,
                  ordered=False).astype(float)

load_net['import'] = cost_kWh * ((load_net['energy_pos'] * 12)/1000)
load_net['export'] = OriginGO.fit * ((load_net['energy_neg'] * 12)/1000)

annual_import = load_net['import'].sum()
annual_export = load_net['export'].sum()

print('import ' + '${:,.2f}'.format(annual_import))
print('export ' + '${:,.2f}'.format(annual_export))

resample24hr = load_net.resample('24h').sum()

print('import ' + '${:,.2f}'.format(resample24hr['import'].mean()/100))
print('export ' + '${:,.2f}'.format(resample24hr['export'].mean()/100))
```

```python
df = load_net.copy()

date = '2019-01-05'
fig, ax = plt.subplots(nrows=1,ncols=1)
df.loc[date].plot(y=['import', 'export'], ax=ax)
df.loc[date].plot.area(y=['energy_pos', 'energy_neg'],
                       ax=ax,
                       secondary_y=True,
                       alpha=0.5)
plt.show()
```

```python


fig, ax = plt.subplots(ncols=1,nrows=1)
df.loc[resample24hr['export'].idxmin().date().strftime('%Y-%m-%d')].plot.area(y='energy', ax=ax,stacked=False)
plt.show()

```
